# Multivariate Plotting With Pandas and Seaborn

```python
# Import Pandas library, loading dataset, and display top 5 rows
import pandas as pd
pokemon = pd.read_csv("Pokemon.csv", index_col=0)
pokemon.head()
```

# Scatter Plot

```python
# Import Seaborn library
import seaborn as sns

# Plotting a scatterplot chart
sns.lmplot(x='Attack', y='Defense', hue='Legendary', 
           markers=['x', 'o'],
           fit_reg=False, data=pokemon)
```

# Box Plot

```python
# Plotting a boxplot chart
sns.boxplot(x="Generation", y="Total", hue='Legendary', data=pokemon)
```

# HeatMap

```python
# Plotting a heatmap chart
sns.heatmap(
    pokemon.loc[:, ['HP', 'Attack', 'Sp. Atk', 'Defense', 'Sp. Def', 'Speed']].corr(),
    annot=True
)
```

# Parallel Coordinates

```python
# Plotting a parallel coordinates chart
from pandas.plotting import parallel_coordinates
p = (pokemon[(pokemon['Type 1'].isin(["Psychic", "Fighting"]))]
         .loc[:, ['Type 1', 'Attack', 'Sp. Atk', 'Defense', 'Sp. Def']]
    )

parallel_coordinates(p, 'Type 1')
```
